Item 1: Regressão Linear

```python
#usar solTrainXtrans e solTestXtrans

# transformando os dados em arrays para a manipulação
x_train = np.array(solTrainXtrans)
y_train = np.array(solTrainY)
x_test = np.array(solTestX)
y_test = np.array(solTestY)

model = LinearRegression().fit(x_train, y_train)
print(f'score: {model.score(x_train,y_train)}')
print(f'intercept: {model.intercept_}')
print(f'coefficients: {model.coef_}')

y_train.shape
```

```python
# Add a column of ones for the intercept
x_train_design = np.column_stack((np.ones(len(x_train)), x_train))
x_test_design = np.column_stack((np.ones(len(x_test)), x_test))

# Calculate the coefficients (X_transposta*X_inversa)^−1*X_transposta*y
Betas = np.linalg.inv(x_train_design.T @ x_train_design) @ x_train_design.T @ y_train

# make predictions
y_pred = x_test_design @ Betas
y_pred.shape
```

```python
# mse = sum((y_test[i] - y_pred[i]) ** 2 for i in range(len(y_test))) / len(y_test)
mean_squared_error(y_train,y_pred) 
```
